```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.neighbors import KNeighborsRegressor
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn import metrics
```

```python
dataset = pd.read_excel('/content/Dados_Criminais_Consolidados_Por_Cem_Mil.xlsx')
```

```python
dataset
```

```python
dataset.info()
```

```python
new_df = pd.DataFrame(columns=['ano', 'cidade', 'populacao', 'numero de casos', 'tipo'])
for i in range(3, 7):
    temp = dataset[['ano', 'cidade', 'populacao']].copy()
    temp['numero de casos'] = dataset[[dataset.columns[i]]]
    temp['tipo'] = dataset.columns[i]

    new_df = pd.concat([new_df, temp])
```

```python
new_df
```

```python
new_df['taxa de criminalidade'] = new_df['numero de casos'] / new_df['populacao']
```

```python
new_df
```

```python
new_df = new_df.drop(['numero de casos'], axis=1)
```

```python
new_df
```

```python
new_df.to_excel("/content/new_dataset.xlsx", index=False, sheet_name ='Sheet1')
```

```python
new_dataset = pd.read_excel("/content/new_dataset.xlsx", sheet_name="Sheet1")
```

```python
new_dataset
```

```python
new_dataset.info()
```

```python
new_dataset.describe()
```

```python
le = LabelEncoder()
```

```python
new_dataset['cidade'] = le.fit_transform(new_dataset['cidade'])
mapping = dict(zip(le.classes_, range(len(le.classes_))))
```

```python
file = open('/content/City_Mapping.txt', 'wt')
for key,val in mapping.items():
    print(str(key) + " - " + str(val) + '\n')
    file.write(str(key) + " - " + str(val) + '\n')
```

```python
ordem_personalizada = ['homicidio', 'furto', 'roubo', 'furto e roubo de veiculo']
mapping_manual = {nome: i for i, nome in enumerate(ordem_personalizada)}
new_dataset['tipo'] = new_dataset['tipo'].map(mapping_manual)
```

```python
with open('/content/Type_Mapping.txt', 'wt') as file:
    for key, val in mapping_manual.items():
        linha = f"{key} - {val}\n"
        print(linha, end='')
        file.write(linha)
```

```python
new_dataset
```

```python
x = new_dataset[new_dataset.columns[0:4]].values
x
```

```python
y = new_dataset['taxa de criminalidade'].values
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=50)
```

```python
x_train
```

```python
y_train
```

MODEL CREATION

Support Vector Machine

```python
model1 = svm.SVR()
model1.fit(x_train, y_train)
y_pred = model1.predict(x_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('R2 score:', metrics.r2_score(y_test, y_pred))
```

Nearest Neighbour

```python
model2 = KNeighborsRegressor(n_neighbors=2)
model2.fit(x_train, y_train)
y_pred = model2.predict(x_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('R2 score:', metrics.r2_score(y_test, y_pred))
```
